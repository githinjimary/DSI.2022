```python
def create_db(db_data):
    from pathlib import Path
    Path(db_data).touch()
    
def establish_db_connection(db_data):
    connection = sqlite3.connect(db_data)
    return connection

def create_and_populate_table(csv_file, connection, db_data):
    c = connection.cursor()
    c.execute('''DROP TABLE IF EXISTS data''')
    c.execute('''CREATE TABLE data ('SNo', 'ObservationDate', 'Province/State', 'Country/Region',
       'Last Update', 'Confirmed', 'Deaths', 'Recovered')''')
    csv_file.to_sql('data', connection, if_exists='append', index = False)
    c.close()
    
```

```python
#LOAD DATA
dff = pd.read_sql('''SELECT * FROM data''',connection)
dff1 = dff.copy()
```

```python
df_confirmed = dff1.groupby(['ObservationDate' , 'Country/Region' ,	'Last Update'])[['Confirmed']].mean()
df_confirmed.reset_index(inplace=True)
df_Deaths = dff1.groupby(['ObservationDate' , 'Country/Region' ,	'Last Update'])[['Deaths']].mean()
df_Deaths.reset_index(inplace=True)
df_Recovered = dff1.groupby(['ObservationDate' , 'Country/Region' ,	'Last Update'])[['Recovered']].mean()
df_Recovered.reset_index(inplace=True)
```

```python
df_Recovered.head(3)
```

```python
#Converting "Observation Date" into Datetime format
df_confirmed["ObservationDate"]=pd.to_datetime(df_confirmed["ObservationDate"])
df_Deaths["ObservationDate"]=pd.to_datetime(df_Deaths["ObservationDate"])
df_Recovered["ObservationDate"]=pd.to_datetime(df_Recovered["ObservationDate"])
df_Deaths.head(3)
```

```python
#get data in cleaned time series format for country
def process_data(data,cntry='Mainland China',window=3):
    df_confirmed = data
    df_confirmed_cntry = df_confirmed[df_confirmed['Country/Region']==cntry]
    final_dataset = df_confirmed_cntry.T[4:].sum(axis='columns').diff().rolling(window=window).mean()[40:]
    df3 = pd.DataFrame(final_dataset,columns=['Total'])
    return df3
```

```python
#get overall wordlwide total for confirmed, recovered and dead cases
def get_overall_total(df3):
    return df3.iloc[:,-1].sum()

conf_overall_total = get_overall_total(df_confirmed)
dead_overall_total = get_overall_total(df_Deaths)
recv_overall_total = get_overall_total(df_Recovered)
print('Overall confirmed:',conf_overall_total)
print('Overall Dead:',dead_overall_total)
print('Overall Recovered:',recv_overall_total)
```

```python
#get total for confirmed, recovered and dead for country 
def get_cntry_total(df3,cntry='Mainland China'):
    return df3[df3['Country/Region']==cntry].iloc[:,-1].sum()

cntry = 'Mainland China'
conf_cntry_total = get_cntry_total(df_confirmed,cntry)
dead_cntry_total = get_cntry_total(df_Deaths,cntry)
recv_cntry_total = get_cntry_total(df_Recovered,cntry)
print(f'{cntry} confirmed:',conf_cntry_total)
print(f'{cntry} Deaths:',dead_cntry_total)
print(f'{cntry} Recovered:',recv_cntry_total)
```

#**LINEGRAPH**

```python
def fig_world_trend(cntry='Mainland China',window=3):
    df3 = process_data(data=df_confirmed,cntry=cntry,window=window)
    df3.head(10)
    if window==1:
        yaxis_title = "confirmed cases"
    else:
        yaxis_title = "confirmed Cases ({}-day MA)".format(window)
    fig = px.line(df3, y='Total', x=df3.index, title='confirmed cases trend for {}'.format(cntry),height=600,color_discrete_sequence =['maroon'])
    fig.update_layout(title_x=0.5,plot_bgcolor='#F2DFCE',paper_bgcolor='#F2DFCE',xaxis_title="Date",yaxis_title=yaxis_title)
    return fig

```

```python
colors = {
    'background': '#111111',
    'bodyColor':'#F2DFCE',
    'text': '#7FDBFF'
}
def get_page_heading_style():
    return {'backgroundColor': colors['background']}


def get_page_heading_title():
    return html.H1(children='COVID-19 spread Dashboard',
                                        style={
                                        'textAlign': 'center',
                                        'color': colors['text']
                                    })

def get_page_heading_subtitle():
    return html.Div(children='Visualize Covid-19 data generated from sources all over the world.',
                                         style={
                                             'textAlign':'center',
                                             'color':colors['text']
                                         })

def generate_page_header():
    main_header =  dbc.Row(
                            [
                                dbc.Col(get_page_heading_title(),md=12)
                            ],
                            align="center",
                            style=get_page_heading_style()
                        )
    subtitle_header = dbc.Row(
                            [
                                dbc.Col(get_page_heading_subtitle(),md=12)
                            ],
                            align="center",
                            style=get_page_heading_style()
                        )
    header = (main_header,subtitle_header)
    return header
```

#**Country dropdown**

```python
def get_country_list():
    return df_confirmed['Country/Region'].unique()

def create_dropdown_list(cntry_list):
    dropdown_list = []
    for cntry in sorted(cntry_list):
        tmp_dict = {'label':cntry,'value':cntry}
        dropdown_list.append(tmp_dict)
    return dropdown_list

def get_country_dropdown(id):
    return html.Div([
                        html.Label('Select Country'),
                        dcc.Dropdown(id='my-id'+str(id),
                            options=create_dropdown_list(get_country_list()),
                            value='Mainland China'
                        ),
                        html.Div(id='my-div'+str(id))
                    ])
```

```python
def graph1():
    return dcc.Graph(id='graph1',figure=fig_world_trend('Mainland China'))
```

```python
#Generate CARDS for overall numbers
def generate_card_content(card_header,card_value,overall_value):
    card_head_style = {'textAlign':'center','fontSize':'150%'}
    card_body_style = {'textAlign':'center','fontSize':'200%'}
    card_header = dbc.CardHeader(card_header,style=card_head_style)
    card_body = dbc.CardBody(
        [
            html.H5(f"{int(card_value):,}", className="card-title",style=card_body_style),
            html.P(
                "Worlwide: {:,}".format(overall_value),
                className="card-text",style={'textAlign':'center'}
            ),
        ]
    )
    card = [card_header,card_body]
    return card
def generate_cards(cntry='Mainland China'):
    conf_cntry_total = get_cntry_total(df_confirmed,cntry)
    dead_cntry_total = get_cntry_total(df_Deaths,cntry)
    recv_cntry_total = get_cntry_total(df_Recovered,cntry)
    cards = html.Div(
        [
            dbc.Row(
                [
                    dbc.Col(dbc.Card(generate_card_content("Recovered",recv_cntry_total,recv_overall_total), color="success", inverse=True),md=dict(size=2,offset=3)),
                    dbc.Col(dbc.Card(generate_card_content("Confirmed",conf_cntry_total,conf_overall_total), color="warning", inverse=True),md=dict(size=2)),
                    dbc.Col(dbc.Card(generate_card_content("Deaths",dead_cntry_total,dead_overall_total),color="dark", inverse=True),md=dict(size=2)),
                ],
                className="mb-4",
            ),
        ],id='card1'
    )
    return cards
```

```python
#DASH Slider for Moving Average Window
def get_slider():
    return html.Div([  
                        dcc.Slider(
                            id='my-slider',
                            min=1,
                            max=15,
                            step=None,
                            marks={
                                1: '1',
                                3: '3',
                                5: '5',
                                7: '1-Week',
                                14: 'Fortnight'
                            },
                            value=3,
                        ),
                        html.Div([html.Label('Select Moving Average Window')],id='my-div'+str(id),style={'textAlign':'center'})
                    ])
```

```python
#Generate APP layout
def generate_layout():
    page_header = generate_page_header()
    layout = dbc.Container(
        [
            page_header[0],
            page_header[1],
            html.Hr(),
            generate_cards(),
            html.Hr(),
            dbc.Row(
                [
                    dbc.Col(get_country_dropdown(id=1),md=dict(size=4,offset=4))                    
                ]
            
            ),
            dbc.Row(
                [                
                    
                    dbc.Col(graph1(),md=dict(size=6,offset=3))
        
                ],
                align="center",

            ),
            dbc.Row(
                [
                    dbc.Col(get_slider(),md=dict(size=4,offset=4))                    
                ]
            
            ),
        ],fluid=True,style={'backgroundColor': colors['bodyColor']}
    )
    return layout
```

```python
app.layout = generate_layout()
```

```python
#Assign DASH Callbacks
@app.callback(
    [Output(component_id='graph1',component_property='figure'), #line chart
    Output(component_id='card1',component_property='children')], #overall card numbers
    [Input(component_id='my-id1',component_property='value'), #dropdown
     Input(component_id='my-slider',component_property='value')] #slider
)
def update_output_div(input_value1,input_value2):
    return fig_world_trend(input_value1,input_value2),generate_cards(input_value1)
```
